# Creating Heatmap using Correlation function
#### Reducing the number of features by finding those features which are fully correlated

```python
plt.figure(figsize=[19,10], facecolor = 'white')
sb.heatmap(df.corr(),annot=True)
```

# Finding feature with max correlation value
#### We find the feature which has correlation value greater than 0.7

```python
for a in range(len(df.corr().columns)):
    for b in range(a):
        if abs(df.corr().iloc[a,b]) > 0.7 :
            name = df.corr().columns[a]
            print(name)
```

# Removing the feature
#### Dropping the feaure which has correlation value greater than 0.7

```python
new_df = df.drop(name,axis=1)
new_df.head()
```

# Finding number of null values in each feature column

```python
new_df.isnull().sum()
```

# Filling the null values
#### To increase accuracy , we replace null values with mean values using fillna function

```python
new_df.update(new_df.fillna(new_df.mean()))
new_df.head()
```

# Handling categorical variables
#### Processind string is not possible , so we use get_dummies function to convert string to numbers
##### Here , in type_white column , 1 --> white wine , 0 --> red wine

```python
next_df = pd.get_dummies(new_df,drop_first = True)
next_df
```

# Creating best quality column
#### We create best quality column to convert the quality range from 0-10 to 0-1

```python
next_df['best quality'] = [1 if x>=7 else 0 for x in df.quality]
next_df
```

# Creating input and output dataset

```python
x = next_df.drop(['quality','best quality'], axis=1)
x
```

```python
y = next_df['best quality']
y
```

# Splitting Dataset to training and testing 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
print(y.shape, y_test.shape, y_train.shape)
```

```python
print(x.shape, x_test.shape, x_train.shape)
```

# Normalization

```python
#importing module
from sklearn.preprocessing import MinMaxScaler
# creating normalization object 
norm = MinMaxScaler()
# fit data
norm_fit = norm.fit(x_train)
new_xtrain = norm_fit.transform(x_train)
new_xtest = norm_fit.transform(x_test)
# display values
print(new_xtrain)
```

# Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

# Checking Accuracy

```python
from sklearn.metrics import accuracy_score
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy = ",test_data_accuracy)
```

# Predicting using user input

```python
input_data = (6.2,0.66,0.48,1.2,0.029,29,0.9892,3.33,0.39,12.8,1)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction==0):
    print('Bad Quality Wine')
else:
    print('Good Quality Wine')
```
